```python
# Create new columns from start_time and end_time to help explore farther more

# Date columns
df1.start_time = pd.to_datetime(df1.start_time)
df1.end_time = pd.to_datetime(df1.end_time)

# Date columns
df1['start_date'] = df1.start_time.dt.strftime('%Y-%m-%d')
df1['end_date'] = df1.end_time.dt.strftime('%Y-%m-%d')

# Day columns
df1['start_day'] = df1.start_time.dt.strftime('%A')
df1['end_day'] = df1.end_time.dt.strftime('%A')

# Hour columns
df1['start_hour'] = df1.start_time.dt.strftime('%H')
df1['end_hour'] = df1.end_time.dt.strftime('%H')

df1.head()
```

```python
# Convert new columns to datetime

# Date columns
df1.start_date = pd.to_datetime(df1.start_date)
df1.end_date = pd.to_datetime(df1.end_date)

# Day columns
df1['start_day'] = df1['start_day'].astype('category')
df1['end_day'] = df1['end_day'].astype('category')

# Hour columns
df1['start_hour'] = df1['start_hour'].astype('int64')
df1['end_hour'] = df1['end_hour'].astype('int64')

df1.info()
```

```python
# Check for duplicates
df1.duplicated().sum()
```

```python
# Check for users value_counts
print(df1.user_type.value_counts())
print(df1.member_gender.value_counts())
print(df1.bike_share_for_all_trip.value_counts())
```

```python
# Convert user_type and member_gender data_type to category
df1['user_type'] = df1['user_type'].astype('category')
df1['member_gender'] = df1['member_gender'].astype('category')
```

```python
# Convert bike_share_for_all_trip data_type to boolean
df1.bike_share_for_all_trip = (df1.bike_share_for_all_trip == 'Yes')
```

```python
# Change duration_sec to duration_min
df1['duration_min'] = df1['duration_sec']/60
```

```python
# Remove all nulled rows
df1 = df1[df1['member_gender'].isnull() == False]
df1 = df1[df1['start_station_id'].isnull() == False]
df1.info()
```

```python
df2 = df1.drop(columns= ['duration_sec', 'start_time', 'end_time'], inplace= False)
df2.head()
```

```python
# Convert member_birth_year data_type to int64
df2['member_birth_year'] = df2['member_birth_year'].astype('int64')
```

```python
# Convert weekdays into Ordered Categorical in start_day column

df2['start_day'] = pd.Categorical(df2['start_day'], dtype="category", categories=
    ['Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday'],
    ordered=False)
ordered_start_days = df2['start_day'].value_counts().sort_index()
ordered_start_days
```

```python
# Convert weekdays into Ordered Categorical in end_day column
df2['end_day'] = pd.Categorical(df2['end_day'], dtype="category", categories=
    ['Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday',],
    ordered=False)
ordered_end_days = df2['end_day'].value_counts().sort_index()
ordered_end_days
```

```python
cols = df2.columns.tolist()
cols
```

```python
# Reorder columns
df = df2[[ 'duration_min',
           'start_date',
           'end_date',
           'start_station_id',
           'start_station_name',
           'start_station_latitude',
           'start_station_longitude',
           'end_station_id',
           'end_station_name',
           'end_station_latitude',
           'end_station_longitude',
           'bike_id',
           'user_type',
           'member_birth_year',
           'member_gender',
           'bike_share_for_all_trip',
           'start_day',
           'end_day',
           'start_hour',
           'end_hour']]

df.head()
```

```python
# Let's look at the pairwise correlations present between features in the data.

numeric_vars = ['duration_min', 'start_station_id','end_station_id', 
                'bike_id', 'start_hour', 'end_hour', 'age']
categoric_vars = ['user_type', 'member_gender', 
                  'start_day', 'end_day']
```

```python
# correlation plot
plt.figure(figsize = [8, 5])
sb.heatmap(df[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)

# Labeling the axes
plt.title(label = "Ford GoBike's Pairwise Correlations Between Features in The Data.", fontsize=20, loc='center');
plt.show()
```

```python
samples = np.random.choice(df.shape[0], 500, replace = False)
df_samp = df.reindex(samples)

g = sb.PairGrid(data = df_samp, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter)
```

**Obviously, there's a strong correlation between the start and end hours, and a weak correlation between start and end stations.**

```python
# compute the logarithm of price to make multivariate plotting easier
def log_trans(x, inverse = False):
    """ quick function for computing log and power operations """
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

df['duration_min_log'] = df['duration_min'].apply(log_trans)
```

```python
# plot the categorical variables against duration_min and member_age again, this time
# with full data and variable transforms
fig, ax = plt.subplots(ncols = 2, nrows = 4 , figsize = [15,10])
for i in range(len(categoric_vars)):
    var = categoric_vars[i]
    sb.violinplot(data = df, x = var, y = 'duration_min_log', ax = ax[i,0], color = base_color);
    ax[i,0].set_yticks(log_trans(np.array([1, 2, 5, 10, 20, 50, 100])));
    ax[i,0].set_yticklabels([1, 2, 5, 10, 20, 50, 100]);
    sb.violinplot(data = df, x = var, y = 'age', ax = ax[i,1], color = base_color);
```

**The use a violin plot and variable transformations reveals much similarity between the categorical variables in relation to age and duration except for gender vs. age. Also, We can see a perfect correspond between start day and end day in relation to ages and duration (min) in a log scale.**

```python
def hist2dgrid(x, y, **kwargs):
    """ Quick hack for creating heat maps with seaborn's PairGrid. """
    palette = kwargs.pop('color');
    bins_x = np.arange(df['age'].min(), df['age'].max()+1, 1);
    bins_y = np.arange(0, 2, 0.1);
    plt.hist2d(x, y, bins = [bins_x, bins_y], cmap = palette, cmin = 0.5);
    plt.yticks(log_trans(np.array([1, 2, 4, 8, 16, 32, 64])), [1, 2, 4, 8, 16, 32, 64]);
```
